```python
import stanza
stanza.download('en')
nlp = stanza.Pipeline('en')
```

```python
import os
import json
import pandas as pd

ANNOTATIONS_PATH = '/content/data/annotations/'
STORIES_PATH = '/content/data/stories/'
annotations = []
stories = []

for fajl in os.scandir(ANNOTATIONS_PATH):
  if fajl.is_file():
    with open(fajl, encoding='utf8') as f:
      annotation = f.read()
      annotations.append(json.loads(json.dumps(annotation)))

for fajl in os.scandir(STORIES_PATH):
  if fajl.is_file():
    with open(fajl, encoding='utf8') as f:
      story = f.read()
      stories.append(story)
      doc = nlp(story)
      uniques = set([ent.text.lower() for ent in filter(lambda e: e.type == "PERSON", doc.ents)])
```
